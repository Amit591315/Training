# Fibonacci Series (Iterative)
This notebook generates Fibonacci numbers using a simple iterative approach.

```python
def fibonacci_series(count: int) -> list[int]:
    """Return the first count Fibonacci numbers using iteration."""
    if count <= 0:
        return []

    series = []
    a, b = 0, 1
    for _ in range(count):
        series.append(a)
        a, b = b, a + b
    return series

count = 10
print("Fibonacci series:")
print(" ".join(str(num) for num in fibonacci_series(count)))
```
